# Introduction

This project was performed to identify whether there was a correlation between number of pages (num_pages), number of ratings (ratings_count), number of review (text_reviews_count) and the average rating (average_rating) of a book. Pandas was used in combination with statsmodels in order to perform the analysis and come to an conclusion.

```python
import pandas as pd
import statsmodels.api as sm
```

```python
books = pd.read_csv('Module 6-books.csv', index_col = 0)
```

```python
books.head()
books.shape
#13719, 9

books.describe()
```

```python
books['authors'].describe()
```

```python
a = books['average_rating']
n = books['num_pages']

model = sm.OLS(a, n).fit()
summary = model.summary()
summary
```

```python
a = books['average_rating']
r = books['ratings_count']

model = sm.OLS(a, r).fit()
summary = model.summary()
summary
```

```python
a = books ['average_rating']
t = books['text_reviews_count']

model=sm.OLS(a, t).fit()
summary = model.summary()
summary
```

## Summary

Based on data collected, it was determined that there are 13,719 total books in the document and of those, 7,605 are unique. The average rating for books is 3.93, the average number of pages is 342, the average ratings count is 17759.09, and average text reviews count is 533. The top author identified was Agatha Christie and appears in the list 69 times.
Based on the results from the OLS regression and examining the P-value of 0.000, we have failed to reject the null hypothesis that average ratings is independent of number of pages. There is a possible correlation between how many pages a book contains and the average rating the book receives. In addition, when examining the P-value (0.000) for the relationship between average rating of a book and number of ratings, it also suggests that there is a possible correlation between the average rating of a book and the number of ratings it receives. Additionally, the P-value (0.000) when examining the average rating of a book and total number of written text reviews, also suggests that there is a correlation between these two variables. 
Overall, it appears there is relationship between the average rating the book receives, the number of written text reviews, number of pages, and average rating.
